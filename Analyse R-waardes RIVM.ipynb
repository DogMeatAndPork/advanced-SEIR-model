```python
# importeert belangrijke functionaliteit
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# importeert de excel gegevens
data = pd.read_excel(r'F:\gebruiker\Documents\R_WAARDES.xlsx')
r_waarde = pd.DataFrame(data, columns=['Column1.Date', 'Column1.Rt_low', 'Column1.Rt_avg', 'Column1.Rt_up',
                                       'Column1.population'])

# dit deel creeert een extra week col. met de corresponderende week van de gegeven dag
r_waarde['Column1.Date'] = r_waarde['Column1.Date'].apply(lambda x: pd.Timestamp(x).strftime('%d-%m-%Y'))
r_waarde['Column1.Date'] = pd.to_datetime(r_waarde['Column1.Date'], format='%d-%m-%Y')
r_waarde['Week_number'] = r_waarde['Column1.Date'].dt.isocalendar().week
print(r_waarde.head())

# een wilde grafiek van R-waarde over tijd
sns.lineplot(data=r_waarde, x="Column1.Date", y="Column1.Rt_avg")
plt.show()


r_waarde_melted = r_waarde.drop(['Week_number', 'Column1.population'], axis=1)
r_waarde_melted = r_waarde_melted.melt("Column1.Date", var_name="R_verdeling", value_name="R_waarde")
print(r_waarde_melted.tail())
ax = sns.lineplot(data=r_waarde_melted, x="Column1.Date", y='R_waarde')
plt.axhline(y=1, color='r', linestyle='--')
ax.set(ylim=(0, None))
ax.set(xlim=('2020-02-17', None))
ax.text('2020-03-09', 3, "Maatregel 1", horizontalalignment='left', size='medium', color='black', weight='semibold')
plt.axvline('2020-03-09', 0, 10)
ax.text('2020-03-16', 4.5, "Lockdown", horizontalalignment='left', size='medium', color='black', weight='semibold')
plt.axvline('2020-03-16', 0, 10)
ax.text('2020-03-23', 3.5, "Aanscherping", horizontalalignment='left', size='medium', color='black', weight='semibold')
plt.axvline('2020-03-23', 0, 10)
ax.text('2020-06-01', 4, "versoepeling", horizontalalignment='left', size='medium', color='black', weight='semibold')
plt.axvline('2020-06-01', 0, 10)
ax.text('2020-07-01', 3, "versoepeling 2", horizontalalignment='left', size='medium', color='black', weight='semibold')
plt.axvline('2020-07-01', 0, 10)
ax.text('2020-08-18', 4.5, "verscherping", horizontalalignment='left', size='medium', color='black', weight='semibold')
plt.axvline('2020-08-18', 0, 10)
ax.text('2020-09-18', 3.5, "verscherping", horizontalalignment='left', size='medium', color='black', weight='semibold')
plt.axvline('2020-09-18', 0, 10)
ax.text('2020-09-29', 4, "verscherping", horizontalalignment='left', size='medium', color='black', weight='semibold')
plt.axvline('2020-09-29', 0, 10)
ax.text('2020-10-05', 3, "mondkapjes", horizontalalignment='left', size='medium', color='black', weight='semibold')
plt.axvline('2020-10-05', 0, 10)
ax.text('2020-10-14', 4.5, "horeca sluit", horizontalalignment='left', size='medium', color='black', weight='semibold')
plt.axvline('2020-10-14', 0, 10)
ax.text('2020-11-04', 3.5, "sluiting", horizontalalignment='left', size='medium', color='black', weight='semibold')
plt.axvline('2020-11-04', 0, 10)
plt.show()
```
